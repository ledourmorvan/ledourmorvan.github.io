## I - Principe de l’ADSL : multiplexage fréquentiel

Source : http://igm.univ-mlv.fr/~dr/XPOSE2006/Julien_Walspurger/principe/principe.html

Le but de l’ADSL (Asymmetric Digital Subscriber Line) est de permettre le transport des données concernant la voix (c’est-à-dire le téléphone, dit RTC) et celles en provenance et à destination d'Internet sur le même support physique : la paire de cuivre. Pour ce faire, l’ADSL utilise le multiplexage fréquentiel. C’est-à-dire que la bande passante du
support va être divisée en trois largeurs de fréquences différentes :
* le premier canal contient les données voix ;
* le second les données à destination d’Internet ;
* le troisième les données en provenance d’Internet.

L’ADSL étant une technologie asynchrone, le débit « abonné vers opérateur » n’est pas le même que le débit « opérateur vers abonné ». En l’occurrence en ADSL, le débit ascendant est plus petit que le débit descendant. Ce choix est dû au fait que sur Internet, l’abonné reçoit plus de données qu’il n’en émet.

**Le multiplexage**

Le multiplexage le plus couramment utilisé pour l’ADSL est le FDM
(Frequency Division Multiplexing) : un multiplexage fréquentiel. Le principe consiste à séparer les plages de fréquences d’une bande passante pour ainsi transiter un type de données par plages. Dans le cas de l’ADSL, la bande passante du support de transmission
utilisée est de 1,1 MHz. Il faut alors la diviser en 3 parties :
* 0 à 4 kHz : la bande de fréquence consacrée aux données voix ;
* 5 à 20 kHz : une bande de fréquence laissée vide pour séparer les données téléphoniques et les données Internet ;
* 20 à 140 kHz : la bande de fréquence consacrée aux données montantes ;
* 150 à 1100 kHz : la bande de fréquence consacrée aux données descendantes.

La modulation La modulation utilisée pour l’ADSL est DMT (Discrete Multi Tone).
Le principe est de découper la bande passante en 256 canaux de 4,3 kHz.
* Le premier canal est réservé aux données voix
* 17 canaux sont réservés aux données numériques montantes
* 225 canaux sont réservés aux données numériques descendantes ;
Ces chiffres montrent clairement pourquoi l’ADSL propose des débits asymétriques et que le débit descendant est plus important que celui montant.

### B : Etude du filtre
**Préparation**
1. Quel type de filtre est nécessaire a priori pour récupérer un unique canal ? Comment choisir sa fréquence caractéristique ? et son facteur de qualité ?

Réponse : un filtre passe-bande, de fréquence caractéristique $f_0$ au milieu de la bande et de facteur de qualité $$Q=\frac{f_0}{\Delta f}$$ avec $\Delta f=4,3$ kHz.  

On va s’intéresser au filtre de Rausch suivant :

![](filesystem:/Rausch.jpeg)
<p style="text-align: center;">Figure n°3 : filtre de Rausch</p>

Il est constitué de résistances, de condensateurs et d’un composant appelé ALI, ici en fonctionnement linéaire et dont la description est donnée en Annexe.

2. Par une analyse basses et hautes fréquences du montage, déterminer le type de filtre réalisé par ce circuit (on lira l’annexe et on appliquera dans chaque cas une loi des mailles adaptée).

Réponse : en basse fréquence, $v_s=v_B=0$ ; en haute fréquence, $i_{R_3}=0$, donc $v_s=v_A=v_B=0$. C'est donc un filtre passe-bande.

On se place désormais dans le cas où $C_1 = C_2 = C$ et $R_1=R_2=R$. 
La valeur de $R_3$ est en revanche différente de celle de $R$. Dans cette situation, la fonction de transfert de ce filtre s’écrit :
$$\underline{H}(\jmath\omega)=-\frac{R_3}{2R}\frac{RC\jmath \omega}{1+\jmath RC\omega +\frac{R_3R}{2}C^2\left(\jmath\omega\right)^2} \ .$$
On choisit $R = 1$ k$\Omega$, $R_3 = 100$ k$\Omega$ et $C = 10$ nF.

3. Quel type de filtre obtient-on ainsi ? Donner l’expression sa fréquence de résonance, de son facteur de qualité et de son gain à la résonance. Faire les applications numériques

Réponse : il s'agit d'un filtre passe-bande de fréquence de résonance
$$f_0=\frac{1}{2\pi}\sqrt{\frac{2}{RR_3C^2}}=\frac1C\sqrt{\frac{1}{2\pi^2RR_3}}=2,251\mbox{ kHz}\  ,$$
de facteur de qualité :
$$Q=\frac{1}{RC\omega_0}=  \sqrt{\frac{R_3}{2R}}\simeq 7\ $$
et de gain à la résonance
$$H_0=-\frac{R_3}{2R}=50\ .$$

4. Quel est l’intérêt d’un tel filtre par rapport à un filtre RLC usuel ?

Réponse : il y a amplification des signaux dans la bande passante.

5. Sachant que la tolérance sur les valeurs de chaque composant est de $1\%$, quelle est l’incertitude que nous avons sur la fréquence de résonance ?

Réponse : on trouve 
$$\frac{u(f_0)}{f_0}=\sqrt{\frac14  \left(\frac{u(R)}{R}\right)^2+\frac14  \left(\frac{u(R_3)}{R_2}\right)^2+\left(\frac{u(C)}{C}\right)^2}\ .$$
Avec $u(C)=\frac{C}{100\sqrt 3}$, $u(R)=\frac{R}{100\sqrt 3}$ et
$u(R_3)=\frac{R_3}{100\sqrt 3}$, 
$$\frac{u(f_0)}{f_0}=\frac{1}{100\sqrt 2}\ ,$$
d'où $u(f_0)=0,016$ kHz.

## Annexe : amplificateur linéaire intégré

L’amplificateur linéaire intégré est un composant électronique permettant, entre autres, d’amplifier la différence de potentiel $v^+-v^-$ entre les entrées $+$ et $−$.

![](filesystem:/ALI.jpeg)
<p style="text-align: center;">Figure n°5 : Amplificateur Linéaire Intégré</p>

En fonctionnement idéal, un ALI vérifie les conditions suivantes :
$$i^+ = i^- = 0$$
et
$$V^+ = V^-$$
en régime linéaire avec $|V_s| < V_{sat}\simeq 12$ V.

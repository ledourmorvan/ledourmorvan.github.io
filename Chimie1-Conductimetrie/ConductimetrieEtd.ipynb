## I – Principe de la conductimétrie

Un conductimètre mesure la conductance $G$ de la portion d’électrolyte comprise entre deux plaques de platine (la cellule conductimétrique) de surface $S$, distantes de $\ell$, plongeant dans la solution :
$$G=\frac1R=\frac{\gamma S}{\ell}\ .$$
Cette formule, établie au chap. 28, est utilisée pour une cellule cylindrique, mais elle n’est qu’approchée, en raison des effets de bords (les lignes de courant ne sont pas toutes parallèles entre elles et
certaines « sortent » de la cellule, mais la proportionnalité de $G$ à 
$\gamma$ est souvent suffisante et
$$G = k\gamma\ ,$$
où $k$ désigne la constante de cellule.

Afin d’accéder à $\gamma$, la cellule de conductimétrie doit être étalonnée : cela consiste à mesurer la conductance $G$ d’une solution étalon de conductivité $\gamma$ connue. L’appareil ajuste alors la valeur de $k$ et affiche $G/k = \gamma$.

Dans la solution, la conductivité est assurée par les ions notés A$_i^{z_i+}$ ou A$_i^{z_i-}$ (par exemple Fe$^{2+}_{(aq)}$, Na$^+_{(aq)}$, SO$_{4(aq)}^{2-}$) ; $z_i$ est donc le nombre algébrique de charges $e$. La formule de Kohlraush (chap. 28) s’écrit alors :
$$\gamma= \sum_i C_i \lambda_i\ ,$$
avec $C_i$ en  **mol.m$^{-3}$**. La grandeur positive $\lambda_i$ est appelée *conductivité ionique molaire*. En réalité $λ_i$ dépend de la concentration totale $C$ en ions dans la solution, selon la loi
empirique
$$\lambda_i= \lambda_i^\circ-\alpha \sqrt C\ .$$

Pour des solutions suffisamment diluées ($C < 10^{-3}$ mol.L$^{-1}$), 
on admet que $\lambda_i \simeq λ_i^\circ$, où $\lambda_i^\circ$ désigne
la *conductivité ionique molaire limite*. Les conductivités ioniques molaires limites sont regroupées dans des tables ramenées à une mole 
de charge : on tabule ainsi
$$\frac{1}{|z_i|}\lambda_i^\circ(\mbox{A}_i^{z_{i}+})=\lambda_i^\circ\left(\frac{1}{|z_i|}\mbox{A}_i^{z_{i}+}\right)\ .$$

Pour les valeurs données dans le tableau ci-dessous,

| $\mbox{ion}$ | $\mbox{H}^+_{(aq)}$  | $\mbox{Na}^+_{(aq)}$  | $\frac13\mbox{Fe}^{3+}_{(aq)}$  | $\mbox{OH}^{-}_{(aq)}$ | $\mbox{CH}_3\mbox{CO}_{2(aq)}^-$ | $\frac12\mbox{SO}^{2-}_{4(aq)}$ |
| :---: | :-----: | :-----: | :-----: | :-----: | :-----: | :-----: |
| $\lambda^\circ\mbox{ (en mS.m$^2$.mol$^{-1}$)}$ | $35,\!0$ | $5,\!0$ | $6,\!8$ | $19,\!9$ | $4,\!1$ | $8,\!0$ |

On utilisera donc la relation
$$\gamma = \sum_i|z_i|c_i\lambda^\circ_i\left(\frac{1}{|z_i|}\mbox{A}_i^{z_{i}+}\right)\ .$$

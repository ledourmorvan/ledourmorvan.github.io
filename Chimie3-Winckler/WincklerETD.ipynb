## I – Principe

### A – But de la manipulation

Le dosage du dioxygène dissous dans l’eau revêt une importance pratique très grande. La mesure de la quantité de dioxygène dissous dans l’eau des rivières, l’eau potable ou de rejets industriels est en relation étroite avec les problèmes de pollution. Il existe une grande variété de procédés physico-chimiques permettant le dosage du dioxygène dissous
dans l’eau. La méthode la plus ancienne a été proposée par Winkler en 1888. Elle continue à être utilisée pour l’analyse des eaux ou pour l’étalonnage des sondes oxymétriques.

### B – Principe

En milieu basique, le dioxygène dissous dans l’eau oxyde Mn(OH)$_{2(s)}$ en Mn(OH)$_{3(s)}$. 
En milieu acide, Mn(OH)$_{3(s)}$ libère des ions Mn$^{3+}_{(aq)}$ qui oxydent les ions iodure I$^-_{(aq)}$ en diiode I$_{2(aq)}$. Le diiode 
I$_{2(aq)}$ est réduit par les ions thiosulfate S$_2$O$^{2-}_{3(aq)}$ 
ce qui permet son dosage.

**Données** (à 298 K)
*  $E_1^\circ\mbox{(S$_4$O$_{6(aq)}^{2-}$/S$_2$O$_{3(aq)}^{2-}$)}=0,08\mbox{ V}$ ; 
* $E_2^\circ\mbox{(I$_{2(aq)}$/I$^{-}_{(aq)}$)}=0,62\mbox{ V}$.

## II – Protocole

Mettre en oeuvre le protocole suivant.

1. Remplir un flacon/erlenmeyer rodé à moitié avec de l’eau du robinet. Y introduire un barreau magnétique ainsi qu’une masse $m =0,6\mbox{ g}$  de chlorure de manganèse II (MnCl$_{2(s)}$) solide puis, à l’aide d’une spatule, quatre pastilles de soude. Remplir le flacon à mi-hauteur du rodage. Le boucher au dessus de l’évier **en évitant d’emprisonner
de l’air**.

**Attention : éviter tout contact de la soude avec la peau. En cas de contact, laver immédiatement avec de l’eau.**

2. Agiter pendant 30 min (avec l’agitateur magnétique). On observe la formation d’un précipité brun. Profiter de ce laps de temps pour étudier la partie théorique.

3. Eteindre l’agitateur, déboucher le flacon et ajouter immédiatement, avec précaution, environ $1\mbox{ mL}$ d’acide sulfurique concentré à l’aide d’une pipette graduée.

4. Introduire ensuite une masse $m = 1\mbox{ g}$ d’iodure de potassium préalablement pesée à l’aide d’une balance. Remettre l’agitateur en fonction, sans reboucher, jusqu’à  l’apparition d’une solution limpide de couleur jaune brun.

5. Un groupe sur deux prépare une volume $V_2 = 200\mbox{ mL}$ d’une solution de thiosulfate de sodium pentahydraté (masse molaire $M = 248,2 \mbox{ g.mol}^{-1}$ de concentration $C_2 = 0,00625\mbox{ mol.L}^{-1}$ .

6. Prélever $V_0 = 50\mbox{ mL}$ de la solution limpide à l’aide d’une fiole jaugée. Verser cette solution dans un erlenmeyer de $250\mbox{ mL}$ et effectuer le dosage par la solution de thiosulfate de sodium de concentration $C_2$. Un peu avant l’équivalence, vous pouvez ajouter quelques cristaux de thiodène pour améliorer la visibilité de l’équivalence.

7. Recommencer la dernière étape si le temps le permet.
